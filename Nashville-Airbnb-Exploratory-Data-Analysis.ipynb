# Nashville Airbnb Exploratory Data Analysis

## Import Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Import and Load Nashville Airbnb Data

```python
listings =  pd.read_csv('C:/Users/New/Downloads/datasets/NashvilleAirBNB/NashvilleListings.csv')
reviews = pd.read_csv('C:/Users/New/Downloads/datasets/NashvilleAirBNB/NashvilleReviews.csv')
neighborhoods = pd.read_csv('C:/Users/New/Downloads/datasets/NashvilleAirBNB/NashvilleNeighbourhoods.csv')

print(f"The listings dataset has {listings.shape[0]} rows and {listings.shape[1]} columns.")
print(f"The reviews dataset has {reviews.shape[0]} rows and {reviews.shape[1]} columns.")
print(f"The neighborhoods dataset has {neighborhoods.shape[0]} rows and {neighborhoods.shape[1]} columns.")
```

## Merge Individual Datasets and Check Overall Information About Data

```python
# Rename the Primary Key Linking Both Datasets as 'id' in the Reviews Dataset
reviews.rename(columns = {'listing_id' : 'id'}, inplace = True)
# Merge All Datasets into One File and Check Data in Preparation for Data Cleaning
nashville = pd.merge(reviews, listings, on='id')
```

```python
# Check General Health of Full Dataset (Overall Info and Missing Values)
nashville.info()
```

## Feature Engineering and Transformation of Data

```python
# Extract Property Type and Listing Rating from Name
nashville['property'] = nashville['name'].str.split('★', expand = True)[0].str.split('·', expand = True)[0].str.split(expand = True)[0]
property_map = {'Rental': 'Rental Unit', 'Guest': 'Guest Suite', 'Bed':'Bed and Breakfast','Tiny': 'Tiny Home', 
                'Farm': 'Farm Stay', 'Religious': 'Religious Building', 'Earthen': 'Earthen home', 'Place': 'Place to Stay',
                'Vacation': 'Vacation Home', 'Shipping': 'Shipping Container','Boutique':'Boutique Hotel', 
                'Serviced': 'Serviced Apartment', 'Boutique': 'Boutique Hotel'}
nashville['property'].replace(property_map, inplace = True)

nashville['rating'] = nashville['name'].str.split('★', expand = True)[1].str.split('·', expand = True)[0].str.strip()
nashville['rating'].replace({'New': '0'}, inplace = True)

# Extract Number of Bedrooms, Beds and Bathrooms from Name and Create Column for Bathroom Type - Either Private or Shared
nashville['bedrooms'] = nashville['name'].str.split('★', expand = True)[1].str.split('·', expand = True)[1].str.split(expand = True)[0]
nashville['bedrooms'].replace({'Studio': '0'}, inplace = True)
nashville['beds'] = nashville['name'].str.split('★', expand = True)[1].str.split('·', expand = True)[2].str.split(expand = True)[0]
nashville['bathrooms'] = nashville['name'].str.split('★', expand = True)[1].str.split('·', expand = True)[3].str.split(expand = True)[0]

# Bathroom Type Transformation
def bath_type(element):
    if element  == 'shared':
        return 'Shared Bath'
    else:
        return 'Private Bath'
nashville['private_bath'] = nashville['name'].str.split('★', expand = True)[1].str.split('·', expand = True)[3].str.split(expand = True)[1].apply(bath_type)

# Split Date Column into Year and Month
nashville['year'] = nashville['date'].str.split('-', expand = True)[0]
nashville['month'] = nashville['date'].str.split('-', expand = True)[1].astype(int)

# Create 'season' Column from 'month' column and Transform 'month' Column
season_map = {1: 'Winter', 2: 'Winter', 3 : 'Spring', 4: 'Spring', 5: 'Spring', 6: 'Summer', 7: 'Summer', 8: 'Summer', 9: 'Fall', 10: 'Fall', 11: 'Fall', 12: 'Winter'}
nashville['season'] = nashville['month'].map(season_map) 
month_map = {1: 'January', 2: 'February', 3 : 'March',4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}
nashville['month'] = nashville['month'].map(month_map)

# Transforming License Column into Boolean
nashville['license'].fillna(0, inplace = True)
def license_func(element):
    if element  == 0:
        return 'No'
    else:
        return 'Yes'
nashville['license'] = nashville['license'].apply(license_func)

# Drop Unneeded Column After Feature Engineering and Data Transformation
nashville.drop(columns = ['date', 'name', 'neighbourhood_group'], inplace = True)

# Check Shape After Feature Engineering is Done
print(f"The merged and cleaned dataset has {nashville.shape[0]} rows and {nashville.shape[1]} columns.")
```

## Insights from Data via Statistics

```python
# How Many Listings are in the  Nashville Airbnb Ecosystem at this time? 
unique_listings = nashville['id'].nunique()
print(f" There are {unique_listings} Airbnb Listings in Nashville, Tennessee as of 2023.")
```

```python
# How Many Hosts are in the  Nashville Airbnb Ecosystem at this time? 
unique_hosts = nashville['host_id'].nunique()
print(f" There are {unique_hosts} Airbnb hosts in Nashville, Tennessee as of 2023.")
```

```python
# How Many Hosts Own Multiple Properties? What Percentage of All Hosts Own Multiple Properties?
multiple_property_hosts = ((nashville.groupby('host_id')['id'].nunique().sort_values(ascending = False)) > 1).sum()
print(f" There are {multiple_property_hosts} Airbnb hosts, who own multiple properties in Nashville, Tennessee as of 2023.")
```

```python
# How Many Hosts Own Multiple Properties? What Percentage of All Hosts Own Multiple Properties?
multiple_property_percentage = round((multiple_property_hosts/unique_hosts) * 100, 2)
print(f"{multiple_property_percentage}% of Airbnb hosts in Nashville, Tennessee own multiple rental properties.")
```

```python
# What Property Types Are Prominent in the Nashville Airbnb Ecosystem? 
print(f"There are {nashville['property'].nunique()} property types in the Nashville Airbnb Ecosystem")
print(f"The 31 property types in the Nashville Airbnb Ecosystem are: {nashville['property'].unique()}")
```

```python
# How Have Property Types Been Added to the Nashville Airbnb Ecosystem Over the Years?
property_type_growth = nashville.groupby('year')['property'].nunique()
sns.barplot(x = property_type_growth.index , y = property_type_growth.values, color = 'blue')
plt.xlabel('Year')
plt.ylabel('Number of Property Types')
plt.xticks(rotation=45)
plt.title('Number of Property Types in the Nashville Airbnb Ecosystem (2009-2023)');
```

```python
# How Many Neighbourhoods Are in the Nashville Airbnb Ecosystem as of 2023?
count_neighbourhood = nashville['neighbourhood'].nunique()
print(f" There are {count_neighbourhood} Neighbourhoods in the Nashville Airbnb ecosystem as of 2023.")
```

## Insights from Data via Visualizations

```python
# Number of Rentals by Year
sns.countplot(data=nashville,x='year', color = 'blue')
plt.xlabel('Year')
plt.ylabel('Total Number of Bookings')
plt.xticks(rotation=45)
plt.title("Number of Nashville Airbnb Bookings by Year (2009 - 2023)");
```

```python
# How have Airbnb Rental Prices Evolved Over the Years
mean_prices = nashville.groupby('year')['price'].mean()
median_prices = nashville.groupby('year')['price'].median()
sns.lineplot(x = mean_prices.index, y = mean_prices.values, color = 'red', label='Mean', marker="o")
sns.lineplot(x = median_prices.index, y = median_prices.values, color = 'blue', label='Median', marker="o")
plt.xlabel('Year')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.title("Prices of Nashville Airbnb Rentals (2009 - 2023)");
```

```python
# How has the Number of Neighbourhoods in the Nashville Airbnb Ecosystem grown since 2009? 
neighbourhood_growth = nashville.groupby('year')['neighbourhood'].nunique()
sns.barplot(x = neighbourhood_growth.index , y = neighbourhood_growth.values, color = 'blue')
plt.xlabel('Year')
plt.ylabel('Total Number of Neighbourhoods')
plt.xticks(rotation=45)
plt.title("Number of Nashville Airbnb Neighbourhoods (2009 - 2023)");
```

```python
# How has the Number of Available Listings in the Nashville Airbnb Ecosystem grown since 2009? 
listings_growth = nashville.groupby('year')['id'].nunique()
sns.barplot(x = listings_growth.index , y = listings_growth.values, color = 'blue')
plt.xlabel('Year')
plt.ylabel('Total Number of Listings')
plt.xticks(rotation=45)
plt.title("Number of Nashville Airbnb Listings (2009 - 2023)");
```

```python
# How has the Numbe of Hosts in the Nashville Airbnb Ecosystem grown since 2009? 
hosts_growth = nashville.groupby('year')['host_id'].nunique()
sns.barplot(x = hosts_growth.index , y = hosts_growth.values, color = 'blue')
plt.xlabel('Year')
plt.ylabel('Total Number of Hosts')
plt.xticks(rotation=45)
plt.title("Number of Nashville Airbnb Hosts (2009 - 2023)");
```

```python
# What kind of Airbnb Listings are Most Rented in the Nashville Area?
listings_type = nashville['room_type'].value_counts()
sns.barplot(x = listings_type.index , y = listings_type.values, color = 'blue')
plt.xlabel('Room Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.title('Number of Nashville Airbnb Booking by Property Type (All Time)');
```
